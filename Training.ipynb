# Come suggerito, il campo inerente il giorno della visita è fondamentale, dunque ho pensato che una buona features da inserire e che possa sostituire le colonne inerenti il giorno della prenotazione e il giorno effettivo della visita fosse il calcolare i giorni mancanti all' effettuazione della visita

```python
from datetime import datetime,timedelta

trainSet['AppointmentDay'] = pd.to_datetime(trainSet['AppointmentDay'],format="%Y-%m-%dT%H:%M:%SZ").dt.date


trainSet['ScheduledDay'] = pd.to_datetime(trainSet['ScheduledDay'],format="%Y-%m-%dT%H:%M:%SZ").dt.date
```

```python
trainSet['GiorniMancanti']=trainSet['AppointmentDay'].sub(trainSet['ScheduledDay'],axis=0)
trainSet['GiorniMancanti']=trainSet['GiorniMancanti'].dt.days.astype('int32')
```

```python
testSet['AppointmentDay'] = pd.to_datetime(testSet['AppointmentDay'],format="%Y-%m-%dT%H:%M:%SZ").dt.date


testSet['ScheduledDay'] = pd.to_datetime(testSet['ScheduledDay'],format="%Y-%m-%dT%H:%M:%SZ").dt.date
```

```python
testSet['GiorniMancanti']=testSet['AppointmentDay'].sub(testSet['ScheduledDay'],axis=0)
testSet['GiorniMancanti']=testSet['GiorniMancanti'].dt.days.astype('int32')
```

# Procedo con l'addestramento e dopo aver provato con diversi tipi di classificatori e tipi di regressione ho scelto il modello che mi ha garantito di raggiungere l'accuratezza più alta ovvero il RandomForest. Questo classificatore è un algoritmo di apprendimento supervisionato che sfrutta  un aggregazione' di decision tree o meglio la previsione effettuata da più alberi. Nello specifico, il risultato finale restituito è la media del risultato numerico restituito dai diversi alberi  i quali singolarmente potrebbero avere una previsione non particolarmente accurata ma facendone una media queste previsioni saranno più vicine al risultato sperato. Tale classificatore è utilizzato per minimizzare il fenomeno dell'overfitting.

```python
train_X = trainSet.drop(columns = ['No-show'])
```

```python
test_X = testSet.drop(columns = ['No-show'])
```

```python
test_Y = testSet["No-show"]
```

```python
modello = RandomForestClassifier(random_state=1)
```

```python
modello.fit(train_X, train_Y)
```

# Calcolo l'accuratezza
